```python
import pandas as pd
import numpy as np
```

```python
# load the data into a df
df = pd.read_csv("../data/routes.txt",header=0,names = ["Airline","AirlineID","Source","SourceID","Destination","DestinationID","Codeshare","Stops","Equipment"])
```

```python
route_id = np.arange(len(df))
for i in range(len(route_id)):
    route_id[i] = str(route_id[i])
len(route_id)
df['RouteID'] = route_id
```

```python
# check to see data makes sense
print(type(df.iloc[0][8]), "This is the first row")
print(df.iloc[0][8])
print(df.iloc[1], "This is the second row")
df
```

```python
# clean the data
df.drop(["Equipment"],axis=1,inplace=True)
df
```

```python
df = df.drop_duplicates(keep="first")
df
```

```python
# TABLE DDLS

# AIRPORTS
CREATE TABLE database1.Airports (
  AirportID VARCHAR(256) NOT NULL,
  Name VARCHAR(256) NULL,
  City VARCHAR(256) NULL,
  Country VARCHAR(256) NULL,
  IATA VARCHAR(3) NULL,
  ICAO VARCHAR(4) NULL,
  TimeZone VARCHAR(256) NULL,
  DST VARCHAR(1) NULL,
  TZ VARCHAR(256) NULL,
  PRIMARY KEY (AirportID));

# AIRLINES
CREATE TABLE database1.Airlines (
  AirlineID` VARCHAR(256) NOT NULL,
  Name VARCHAR(256) NULL,
  IATA VARCHAR(256) NULL,
  ICAO VARCHAR(256) NULL,
  Country VARCHAR(256) NULL,
  Active VARCHAR(256) NULL,
  PRIMARY KEY (AirlineID));

# COUNTRIES
CREATE TABLE database1.Countries (
  Name VARCHAR(256) NOT NULL,
  ISO VARCHAR(45) NULL,
  PRIMARY KEY (Name));

# PLANES
CREATE TABLE database1.Planes (
  Name VARCHAR(256) NOT NULL,
  IATA VARCHAR(45) NULL,
  ICAO VARCHAR(45) NULL,
  PRIMARY KEY (Name));

# ROUTES
CREATE TABLE database1.Routes (
  RouteID INT NOT NULL,
  Airline VARCHAR(256) NULL,
  AirlineID VARCHAR(256) NULL,
  Source VARCHAR(256) NULL,
  SourceID VARCHAR(256) NULL,
  Destination VARCHAR(256) NULL,
  DestinationID VARCHAR(256) NULL,
  Codeshare VARCHAR(256) NULL,
  Stops VARCHAR(256) NULL,
  PRIMARY KEY (RouteID));
```
